```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=1000,
                 n_features=2,
                 centers=4,
                 cluster_std=3.0,
                 random_state=42)
```

# k-means clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
kmeans.labels_
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, 
                              ncols=2, 
                              sharey=True, 
                              figsize=(10,6))
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, 
                              ncols=2, 
                              sharey=True, 
                              figsize=(10,6))

ax1.scatter(data[0][:,0],
           data[0][:,1],
           c=kmeans.labels_,
            cmap ="rainbow")

ax2.scatter(data[0][:,0],
           data[0][:,1],
           c=data[1],
           cmap="rainbow")
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, 
                              ncols=2, 
                              sharey=True, 
                              figsize=(10,6))

ax1.scatter(data[0][:,0],
           data[0][:,1],
           c=kmeans.labels_,
            cmap ="rainbow")

ax1.scatter(x=centers[:,0],#all the rows
           y=centers[:,1])#all the row


ax2.scatter(data[0][:,0],
           data[0][:,1],
           c=data[1],
           cmap="rainbow")
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, 
                              ncols=2, 
                              sharey=True, 
                              figsize=(10,6))

ax1.scatter(data[0][:,0],
           data[0][:,1],
           c=kmeans.labels_,
            cmap ="rainbow")

ax1.scatter(x=centers[:,0],#all the rows
           y=centers[:,1],
           c='black', #center color
           s=100) #size of centroid


ax2.scatter(data[0][:,0],
           data[0][:,1],
           c=data[1],
           cmap="rainbow")
```

```python
sum_square={}

for k in range(1,10):
    kmeans = KMeans(n_clusters=k).fit(data[0]) #k is changing from 1 to 10
    sum_square[k]=kmeans.inertia_
```

```python
sum_square
```

```python
plt.plot(list(sum_square.keys()),
        list(sum_square.values()))
```

```python
plt.plot(list(sum_square.keys()),
        list(sum_square.values()),
    #some figure aesthetics
         linestyle='-', #for continuous line
         marker='H', #for hexagon
         color='g', #for green color
         markersize= 8, 
         markerfacecolor='b') # for blue color
    
```
